```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data1 = pd.read_csv('2015.csv', low_memory=False)
data2 = pd.read_csv('2016.csv', low_memory=False)
data3 = pd.read_csv('2017.csv', low_memory=False)
combined_data = pd.concat([data1, data2, data3], ignore_index=True)
combined_data.head(10)
```
